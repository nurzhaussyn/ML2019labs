```python
########################
## Nurzhaussyn Maksat ##
########################
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('switzerland-4180466_1920.jpg')

print(img.shape)

# cv2.imshow("swit", img)
# cv2.waitKey(0)
```

```python
# Rotate 90 clockwise/counterclockwise
img_rotate_90_clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2.imwrite('swit_rotate_90_clockwise.jpg', img_rotate_90_clockwise)

img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2.imwrite('swit_rotate_90_counterclockwise.jpg', img_rotate_90_counterclockwise)
```

```python
# Flip horizontally
img_flip_lr = cv2.flip(img, 1)
cv2.imwrite('swit_cv_flip_lr.jpg', img_flip_lr)
```

```python
# Translations 
rows, cols = img.shape

M = np.float32([[1,0,200],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('swit',dst)
cv2.waitKey(0)
```

```python
# Canny edge detection
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
#### BGR channel decomposition
cv2.imshow('all',img)
cv2.imshow('blue',img[:,:,0])
cv2.imshow('green',img[:,:,1])
cv2.imshow('red',img[:,:,2])
cv2.waitKey(0)
```
